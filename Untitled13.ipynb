```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
import time
from sklearn.svm import LinearSVC
import pickle
from skimage.feature import hog
from utils import sliding_window, pyramid, non_max_suppression
%matplotlib inline
```

```python

filename = 'E:\\HOG\\finalized_model.sav'
svc = pickle.load(open(filename, 'rb'))
```

```python
image  = cv2.imread('E:\\HOG\\15.jpg')
plt.imshow(image)
```

```python
clone = image.copy()
```

```python
def get_hog_features(img, orient, pix_per_cell, cell_per_block, 
                        vis=False, feature_vec=True):
    """
    Return the hog features of the given input image
    Call with two outputs if vis==True"""
    if vis == True:
        features, hog_image = hog(img, orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell),
                                  cells_per_block=(cell_per_block, cell_per_block), transform_sqrt=True, 
                                  visualise=vis, feature_vector=feature_vec)
        return features, hog_image
    # Otherwise call with one output
    else:      
        features = hog(img, orientations=orient, pixels_per_cell=(pix_per_cell, pix_per_cell),
                       cells_per_block=(cell_per_block, cell_per_block), transform_sqrt=True, 
                       visualise=vis, feature_vector=feature_vec)
        return features
```

```python
detections = []
downscale_power = 0
downscale = 2
window_size = (300,300)
window_step_size = 32
threshold = .4
for im_scaled in pyramid(image, downscale=downscale, min_size=window_size):
    # if the width or height of the scaled image is less than
    # the width or height of the window, then end the iterations
    if im_scaled.shape[0] < window_size[1] or im_scaled.shape[1] < window_size[0]:
        break
    for (x, y, im_window) in sliding_window(im_scaled, window_step_size, window_size):
        if im_window.shape[0] != window_size[1] or im_window.shape[1] != window_size[0]:
            continue
        features = []
        file_features = []
        im_window = cv2.resize(im_window,(100,100))
        feature_image = np.copy(im_window)
        hog_features = []
        hog_features.append(get_hog_features(feature_image ,9, 8, 2,vis=False, feature_vec=True))
        hog_features = np.ravel(hog_features)
        file_features.append(hog_features)
        features.append(np.concatenate(file_features))
        prediction = svc.predict(features)
        if prediction == 0:
            x1 = int(x * (downscale ** downscale_power))
            y1 = int(y * (downscale ** downscale_power))
            detections.append((x1, y1,x1 + int(window_size[0] * (downscale ** downscale_power)),
                               y1 + int(window_size[1] * (downscale ** downscale_power))))
    downscale_power += 1  
        
        
```

```python

clone_before_nms = clone.copy()
for (x1, y1, x2, y2) in detections:
    cv2.rectangle(clone_before_nms, (x1, y1), (x2, y2), (0, 255, 0), thickness=10)
plt.imshow(clone_before_nms)
```

```python
detections = non_max_suppression(np.array(detections), threshold)
clone_after_nms = clone
        # Display the results after performing NMS
for (x1, y1, x2, y2) in detections:
            # Draw the detections
    cv2.rectangle(clone_after_nms, (x1, y1), (x2, y2), (0, 255, 0), thickness=10)
plt.imshow(clone_after_nms)
```
